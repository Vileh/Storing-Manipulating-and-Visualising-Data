a.) **(5 points)** Read the file  **unemployment.csv** into a dataframe called _unemp_ using Pandas 
- Remove any column or row which does not correspond to a country or date
- All invalid values should be given a value of 0
- All columns except the date name should be numeric while the date should be of type datetime
------------------------------------


```python
# YOUR CODE HERE
unemp = pd.read_csv('unemployment.csv')
unemp.drop([0,1], inplace=True)
unemp.loc[:, unemp.columns != 'Date'] = unemp.loc[:, unemp.columns != 'Date'].apply(lambda x: pd.to_numeric(x, errors='coerce')).fillna(0)
unemp['Date'] = unemp['Date'].str.lstrip()
unemp['Date'] = pd.to_datetime(unemp['Date'], format='%b-%Y')
```

```python
unemp
```

------------------------------------
------------------------------------
b.) **(2 points)** Set the index to the Date

------------------------------------


```python
# YOUR CODE HERE
unemp.index = unemp['Date']
unemp.drop('Date', inplace=True, axis=1)
```

c.) **(5 points)** Create a line plot corresponding to Spain and Colombia using complementary colours (please explicitly make the colours complementary)

------------------------------------

```python
# YOUR CODE HERE
plt.plot(unemp['Spain'], c=colorsys.hsv_to_rgb(0.25,1,1))
plt.plot(unemp['Colombia'], c=colorsys.hsv_to_rgb(0.75,1,1))

plt.xticks(np.array(unemp.index), rotation=45);
plt.title('Unemployment rate in Spain and Colombia')
plt.xlabel('Date')
plt.ylabel('Unemployment rate')
plt.legend(['Spain', 'Colombia'])
```


------------------------------------

d.) **(8 points)** Complete the function below which takes in an integer and returns N colours equally spaced in hue (with maximum saturation and value/luminance) as rgb values

------------------------------------


```python
def create_colours(N):
    colours=[(i, 1, 1) for i in np.arange(0, 1.01, 1/(N-1))]
    return colours
```

```python
colors = create_colours(10)
colors
```


------------------------------------

e.) **(10 points)** Plot the unemployment rates in 2020 
- include all countries whose names start with a letter which comes before but **not** including G 
- use the colour function you have just written. 

------------------------------------


```python
# YOUR CODE HERE
selected_columns = [col for col in unemp.columns if col[0].upper() < 'G']
colors = create_colours(len(selected_columns))
colorsys.hsv_to_rgb(0.25,1,1)
for i in selected_columns:
    plt.plot(unemp[i], c=colorsys.hsv_to_rgb(*colors.pop()))

plt.xticks(np.array(unemp.index), rotation=45);
plt.title('Unemployment rate for selected countries')
plt.xlabel('Date')
plt.ylabel('Unemployment rate')
plt.legend(selected_columns)
```
